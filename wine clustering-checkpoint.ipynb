```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(style='darkgrid', font_scale=1.4)
```

```python
data=pd.read_csv('wine-clustering.csv')
data
```

```python
data.isna().sum().sum()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.corr()
```

```python
X=data[["Alcohol","Malic_Acid"]]
X
```

```python
#Scatterplot of X
plt.figure(figsize=(10,6))
sns.scatterplot(x = 'Alcohol',y = 'Malic_Acid',  data = X  ,s = 60 )
plt.xlabel('Alcohol')
plt.ylabel('Malic_Acid') 
plt.title('Malic_Acid vs Alcohol')
plt.show()
```

```python
#Importing KMeans from sklearn
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,11):
    km=KMeans(n_clusters=i)
    km.fit(X)
    wcss.append(km.inertia_)
```

```python
#The elbow curve
plt.figure(figsize=(12,6))
plt.plot(range(1,11),wcss)
plt.plot(range(1,11),wcss, linewidth=2, color="red", marker ="8")
plt.xlabel("K Value")
plt.xticks(np.arange(1,11,1))
plt.ylabel("WCSS")
plt.show()
```

```python
#Taking 5 clusters
km1=KMeans(n_clusters=5)
#Fitting the input data
km1.fit(X)
#predicting the labels of the input data
y=km1.predict(X)
#adding the labels to a column named label
X["label"] = y
#The new dataframe with the clustering done
X.head()
```

```python
#Scatterplot of the clusters
plt.figure(figsize=(10,6))
sns.scatterplot(x = 'Alcohol',y = 'Malic_Acid',hue="label",  
                 palette=['green','orange','brown','dodgerblue','red'], legend='full',data = X  ,s = 60 )
plt.xlabel('Alcohol')
plt.ylabel('Malic_Acid') 
plt.title('malic_acid vs alcohol')
plt.show()
```
